```python
import os

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from tqdm import tqdm

import wandb
```

```python
wandb.login()
```

```python
api = wandb.Api()
project = 'seanxx/11711-hw4-ablation'
runs = api.runs(project)
```

```python
results = []
for run in tqdm(runs):
    if 'pre' in run.tags:
        continue

    config = run.config
    summary = run.summary

    result = {}
    result['id'] = run.id
    result['embedding_path'] = config.get('embedding_path', None)
    result['test_roc_auc'] = summary.get('test_roc_auc', None)
    result['test_pr_auc'] = summary.get('test_pr_auc', None)

    results.append(result)

df = pd.DataFrame(results)
df
```

```python
hue_col = 'Embedding Model'
df[hue_col] = df['embedding_path'].apply(
    lambda x: 'ChatGPT text-embedding-3-large' if 'icd2embedding' in x else 'ClinicalBERT')
```

```python
figsize = (8.5, 7)
```

```python
plt.figure(figsize=figsize)

sns.scatterplot(data=df, x='test_roc_auc', y='test_pr_auc', hue=hue_col)

plt.title('Scatter Plot of Ablation Experiment Performance', fontsize=14)
plt.xlabel('Test ROC-AUC', fontsize=12)
plt.ylabel('Test PR-AUC', fontsize=12)
plt.legend(loc='upper left', fontsize=12, title=hue_col, title_fontsize=12)
plt.xticks(np.linspace(0.62, 0.72, 21), rotation=45)
plt.yticks(np.linspace(0.67, 0.77, 11), rotation=0)
plt.grid(True, linestyle='--')
plt.gca().set_axisbelow(True)

plt.tight_layout()
plt.savefig('scatter_plot.png', dpi=300)
plt.show()
```

```python
plt.figure(figsize=figsize)

best_model_ids = ['36ce4x0h', 'kiz7ll7x']
model_name = ['HINT-ChatGPT-v1', 'HINT-ChatGPT-v2']
colors = ['tab:red', 'tab:blue']

for i, model_id in enumerate(best_model_ids):
    run = api.run(os.path.join(project, model_id))
    hist_df = run.history().iloc[:-1, :]
    hist_df['Epoch'] = hist_df['_step'] + 1
    plt.plot(hist_df['Epoch'], hist_df['train_loss'],
             label=f'{model_name[i]} Train Loss', color=colors[i])
    plt.plot(hist_df['Epoch'], hist_df['val_loss'], linestyle='dashed',
             label=f'{model_name[i]} Validation Loss', color=colors[i])

plt.title('Loss Curves of 2 Best Models', fontsize=14)
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Loss', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--')
plt.gca().set_axisbelow(True)

plt.tight_layout()
plt.savefig('loss_plot.png', dpi=300)
plt.show()
```
